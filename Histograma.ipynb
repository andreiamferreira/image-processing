```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

im1 = cv2.imread('/content/gdrive/MyDrive/Intro. Processamento de Imagens Digitais/Aula 2 - Histograma/Fig0316.tif')

# CVTCOLOR parametros -> qual a variaivel a ser alterada
# transformando o rgb paa uma imagem em escala de cinza
img = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)

# equalizeHist faz a equação do gonzales
#operação de equalização do histograma
h_eq = cv2.equalizeHist(img)

plt.figure()
plt.title("Histograma Original")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(img.ravel(), 256, [0,256])
plt.xlim([0, 256])
plt.show()
plt.figure()
plt.title("Histograma Equalizado")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.hist(h_eq.ravel(), 256, [0,256])
plt.xlim([0, 256])
plt.show()
cv2.waitKey(0)

figure, axarr = plt.subplots(1,2, figsize=(20,20))
axarr[0].imshow(im1, cmap='gray')
axarr[1].imshow(h_eq, cmap='gray')
```
